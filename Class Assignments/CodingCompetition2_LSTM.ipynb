# LSTM Example 3: predict the next word in a sentence

Below is a coding example using an LSTM model in TensorFlow to work with a sequence of strings. The example demonstrates how to process a list of strings and train an LSTM model to predict the next word in a sentence. This is a common task in natural language processing called text generation.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from keras.layers import Dropout
```

```python
# In this example, we will use a simple list of sentences to train an LSTM model to predict the next word based on the previous words in the sequence.
# Sample list of sentences (input data)
sentences = [
    "Kakashi is often seen reading his favorite book, 'Icha Icha Paradise.'",
    "Kakashi is the leader of Team 7, guiding Naruto, Sasuke, and Sakura.",
    "Kakashi possesses the Sharingan, a powerful dōjutsu gifted by Obito Uchiha.",
    "Kakashi created the Chidori, a high-speed lightning-based jutsu for combat.",
    "Kakashi is known for being calm, strategic, and highly intelligent in battle.",
    "After the Fourth Great Ninja War, Kakashi became the Sixth Hokage of Konoha.",
    "Kakashi plays an important role in training Naruto and Sasuke to grow stronger.",
    "Kakashi’s Sharingan was a gift from his teammate Obito during a dangerous mission.",
    "Kakashi specializes in using lightning-style jutsu, especially in fast-paced fights.",
    "Kakashi can summon a pack of ninja dogs to help him track enemies and fight.",
    "Kakashi’s father, Sakumo Hatake, was a respected ninja who died when Kakashi was young.",
    "Kakashi learned the value of teamwork after losing his teammates during missions.",
    "Kakashi played a key role during the Fourth Great Ninja War as a battlefield commander.",
    "Kakashi is proficient in genjutsu, although he mainly excels in ninjutsu and taijutsu.",
    "Kakashi’s full face is a mystery, as he always wears a mask over his mouth.",
    "Kakashi enjoys reading romance novels written by Jiraiya, especially during downtime.",
    "Kakashi is considered a tactical genius, able to outthink opponents with ease.",
    "Kakashi deeply regrets the loss of his teammate Rin Nohara during a critical mission.",
    "Kakashi’s father, Sakumo Hatake, was known as the White Fang, a legendary warrior.",
    "Kakashi is frequently late to meetings, often with excuses like being lost on the 'path of life.'",
    "Kakashi was responsible for training Naruto, Sasuke, and Sakura in teamwork and discipline.",
    "Kakashi is famous for copying over 1,000 jutsu using his Sharingan in battle.",
    "Kakashi cares deeply about his friends and will do anything to protect them.",
    "Kakashi is known as the Copy Ninja due to his ability to replicate any jutsu.",
    "Kakashi initially failed Team 7's first bell test, teaching them the importance of teamwork.",
    "Kakashi believes in the Will of Fire, the philosophy that protects the Hidden Leaf Village.",
    "Kakashi unlocked the Mangekyō Sharingan, which gave him access to the powerful Kamui technique.",
    "Kakashi fought during the Third Great Ninja War, where he earned his reputation as a hero.",
    "Kakashi eventually loses his Sharingan during a fierce battle in the Fourth Great Ninja War.",
    "Kakashi mastered a wide variety of jutsu, making him highly versatile in combat situations.",
    "Kakashi continues to mentor younger ninjas, passing down his wisdom and experience.",
    "Kakashi teaches his students the importance of loyalty and trusting their comrades.",
    "Kakashi often hides his emotions, appearing distant, but he deeply cares for those close to him.",
    "Kakashi’s Mangekyō Sharingan allows him to use Kamui, a space-time manipulation technique.",
    "Kakashi was promoted to Jōnin at a young age, showing his exceptional skill and talent.",
    "Kakashi plays a crucial role in helping Naruto and his friends save the world from destruction.",
    "Kakashi joins forces with Naruto, Sasuke, and Sakura to defeat Kaguya Ōtsutsuki in the war.",
    "Kakashi prefers to avoid unnecessary battles, often seeking peaceful solutions when possible.",
    "Kakashi is a strong supporter of Naruto's dream to become Hokage and brings peace to the village.",
    "Kakashi survived numerous difficult battles throughout his life, proving his resilience and strength."
    
]

# Tokenize the sentences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
total_words = len(tokenizer.word_index) + 1

# Create input sequences and corresponding labels
input_sequences = []
for sentence in sentences:
    token_list = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i + 1]
        input_sequences.append(n_gram_sequence)

# Pad sequences to ensure uniform length
max_sequence_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')

# Split input sequences into features and labels
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

# Convert labels to one-hot encoded format
y = tf.keras.utils.to_categorical(y, num_classes=total_words)

# Build the LSTM model
model = Sequential()
model.add(Embedding(total_words, 256))
model.add(LSTM(100, activation='relu'))
model.add(Dropout(0.2))  # Dropout after second LSTM layer
model.add(Dense(total_words, activation='softmax'))
optimizer = tf.keras.optimizers.Adam(learning_rate=0.005)  # You can also try 0.0005
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X, y, epochs=300, batch_size=128, verbose=1)

# Function to generate text based on seed text
def generate_text(seed_text, next_words, max_sequence_len):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len - 1, padding='pre')
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break
        seed_text += " " + output_word
    return seed_text
```

```python
# Test the text generation - Train
seed_text = "Kakashi is often seen"
next_words = 8
generated_text = generate_text(seed_text, next_words, max_sequence_len)
print(f"Generated Text: {generated_text}")
```

```python
# Test the text generation - Test 1
seed_text = "Kakashi is known as"
next_words = 8
generated_text = generate_text(seed_text, next_words, max_sequence_len)
print(f"Generated Text: {generated_text}")
```
